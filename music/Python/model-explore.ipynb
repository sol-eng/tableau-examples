```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
```

```python
music = pd.read_csv("../data/music-genre-cleaned.csv")
```

```python
X = music.valence.to_numpy().reshape(-1, 1)
y = music.popularity.to_numpy()
```

```python
regr = RandomForestRegressor(max_depth = 2, random_state = 0)
regr.fit(X, y)
```

```python
preds = regr.predict(X)
```
